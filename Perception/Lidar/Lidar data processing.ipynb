```python
import numpy as np
import os
import pickle
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
LIDAR_DATA_FOLDER = "../../Simulation/lidar_data"
```

```python
lidar_data_filenames = os.listdir(LIDAR_DATA_FOLDER)
len(lidar_data_filenames)
```

### Point cloud

```python
point_cloud_filenames = sorted([filename for filename in lidar_data_filenames 
                                if filename.startswith("lidar_point_cloud")], 
                              key=lambda filename: int(filename.split(".")[0].split("_")[-1]))
len(point_cloud_filenames)
```

```python
step = 10 # Simulation step
```

```python
with open(os.path.join(LIDAR_DATA_FOLDER, point_cloud_filenames[step]), "rb") as f:
    point_cloud = pickle.load(f)
```

```python
points_to_visualise = [point[:3] for point in point_cloud]
```

```python
X, Y, Z = zip(*points_to_visualise)
```

```python
fig = plt.figure(figsize=(9, 6))
# Create 3D container
ax = plt.axes(projection = '3d')
# Visualize 3D scatter plot
ax.scatter3D(X, Y, Z)
# Give labels
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
# Save figure
plt.savefig('3d_scatter.png', dpi = 300);
```
