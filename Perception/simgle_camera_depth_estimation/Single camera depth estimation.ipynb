### Detect cone

```python
def detect_cones(image):
    # So far we are not using NN inference, just return constant answer
    return [
        (330, 190, 260, 275) # x1, y1, x2, y2 (1 is bottom left, 2 is top right)
    ]
```

```python
def draw_bbox(image, bboxs):
    # Create figure and axes
    fig, ax = plt.subplots()

    # Display the image
    ax.imshow(image)

    # Create a Rectangle patch
    for box in bboxs:
        bottom_left = box[:2]
        width = box[2] - box[0]
        height = box[3] - box[1]
        print(bottom_left)
        rect = patches.Rectangle(bottom_left, width, height, linewidth=1, edgecolor='r', facecolor='none')

        # Add the patch to the Axes
        ax.add_patch(rect)

    plt.show()
```

```python
bboxs = detect_cones(image)
print(f"Detected {len(bboxs)} cones")
```

```python
draw_bbox(image, bboxs)
```
