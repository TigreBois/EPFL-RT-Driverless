```python
import csv
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
# All tracks

tracks_files = glob.glob('*.csv')

tracks = {}

for track_file in tracks_files:
    with open(track_file, newline='') as track:
        values = []
        xs = []
        ys = []
        colors = []
        for cone in track:
            reader = csv.reader(track, delimiter=',')
            for row in reader:
                xs.append(float(row[0]))
                ys.append(float(row[1]))
                colors.append(row[2])
        values = [xs, ys, colors]
        tracks[f"{track_file}"] = values
        
f, axs = plt.subplots(len(tracks),1)

for i, tr in enumerate(tracks):
    axs[i].set_title(tr)
    axs[i].scatter(tracks[tr][0],tracks[tr][1], color=tracks[tr][2])
    axs[i].figure.set_figheight(20)
    axs[i].set_aspect('equal', adjustable='box', anchor='C')
    #axs[i].transData.transform([(0,1),(1,0)])-axs[i].transData.transform((0,0))
```
